### 03. Data quality and consistency checks

```python
# Checking for duplicates
df_customers[df_customers.duplicated()]
```

```python
# Finding missing values
df_customers.isnull().sum()
```

```python
# Checking missing values in first_name column
df_nan = df_customers[df_customers['first_name'].isnull() == True]
```

```python
df_nan
```

```python
# Addressing missing values - drop the column 'first_name'
df_customers_dropfn = df_customers.drop(columns=['first_name'])
```

```python
# Checking the result
df_customers_dropfn.head()
```

```python
# Checking for duplicates df without the first_name column
df_customers_dropfn[df_customers_dropfn.duplicated()]
```

```python
# Check for mixed types
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```

```python
# Checkin the data type
df_customers['first_name'].dtype
```

```python
# Addressing mixed-type data : changing 'first_name' column into string type
df_customers['first_name'] = df_customers['first_name'].astype('str')
```

```python
# Checking the result
df_customers.isnull().sum()
```

### 04. Combine customer data with the rest prepared Instacart data.

```python
# Import prepared Instacart data
df_ords_prods = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_frequent_customer.pkl'))
```

```python
# Checking the data
df_ords_prods.head()
```

```python
df_ords_prods.shape
```

```python
df_customers.shape
```

```python
# Checking the data type 'user_id' as their shared column for the both dataframes
df_ords_prods['user_id'].dtype
```

```python
df_customers['user_id'].dtype
```

```python
# Changing data type for df_customers['user_id'] into int32 to be the same data type as df_ords_prods['user_id']
df_customers['user_id'] = df_customers['user_id'].astype('int32')
```

```python
# Checking the output
df_customers['user_id'].dtype
```

```python
# Merging data frames
df_merged = df_ords_prods.merge(df_customers, on = 'user_id')
```

```python
# Checking the output
df_merged.head()
```

```python
df_merged.shape
```

### 05. Changing column 'prices' from the practice example

```python
# Checking stats for column 'prices'
df_merged['prices'].describe()
```

```python
#Set values >100 as NaN / missing
df_merged.loc[df_merged['prices'] >100, 'prices'] = np.nan
```

```python
# Checking the output 'prices' stats
df_merged['prices'].describe()
```

### 06. Exporting the merged file
