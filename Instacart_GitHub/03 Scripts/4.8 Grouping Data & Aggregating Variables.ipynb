### 06. Deriving Columns with loc()

```python
#5 check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer)
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','max','min']})
```

```python
#6 Create a new column for average price spent for each user_id
ords_prods_merge['avg_price_spent'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Derriving a column with loc()
ords_prods_merge.loc[ords_prods_merge['avg_price_spent'] < 10, 'spender_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_price_spent'] >= 10, 'spender_flag'] = 'High spender'
```

```python
# Checking the output
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head(15)
```

```python
#7 Create a new column for the median in the “days_since_prior_order” column
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Derriving a column with loc()
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Checking the output
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
